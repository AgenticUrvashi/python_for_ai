**que01:** 1. Ek program: number lo, print karo "Positive", "Negative", ya "Zero".



```python
nums=int(input("enter number"))

if nums>0:
    print("positive")
elif nums<0:
    print("negative")
else:
    print("zero")
```

**que02:** Ek program: age lo. 18+ → "You can vote", warna → "Too young".

```python
age = int(input("enter age: "))

if age >= 18:
    print("you can vote")
elif age <= 18:
    print("too young")
```

**que03:** Marks lo aur upar wala grade A/B/C/D print karo.


```python
marks=input("enter your marks to get grade")

if marks >= str(90):
    print("grade A")
elif marks >= str(70):
    print("grade B")
elif marks >= str(50):
    print("grade C")
elif marks < str(50):
    print("grade D")
```
