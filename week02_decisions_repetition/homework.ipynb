**que 03:** Number guessing game khelo aur jeeto!


```python
import random

secret = random.randint(1, 100)     

attempts = 0

print("=== Guess the number (1-100) ===")

while True:

    guess = int(input("Your guess: "))

    attempts = attempts + 1

    if guess < secret:

        print("Too low! ⬆️")

    elif guess > secret:

        print("Too high! ⬇️")

    else:

        print(f"Correct! 🎉 You got it in {attempts} attempts")

        break
```
