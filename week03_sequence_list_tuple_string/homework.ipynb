**que02:** 3 students ka naam aur 3 subjects ke marks rakho; har student ka total print karo.


```python
list = [

    ["shweta",89,94,84],
    ["janu",67,82,90],
    ["ravi",78,87,65],

]

for row in list:
        print(f"{row[0]} : total = {sum(row[1:5])}", end=" ")

        print()
```

**que03:** Ek 3x3 grid ke saare items nested loop se print karo.


```python
grid = [
    [45,78,67],
    [56,87,91],
    [34,67,89],
]

for row in grid:
    for col in row:
        print(col, end=" ")

    print()
```
